```python
#导入模块
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Line
%matplotlib inline

#读取数据
data=pd.read_excel('本市去年部分统计数据.xls',sheet_name='生产总值',skiprows=2)
#提取x轴标签
items=list(data['生产总值'][0:])
#提取y轴数据
y_data1=list(data['前三季度'][0:])
y_data2=list(data['增长（％）'][0:])
#调用Line()方法绘制柱状图
line=Line()
line.add_xaxis(items)
line.add_yaxis("前三季度",y_data1)
line.add_yaxis("增长（％）",y_data2)
line.set_global_opts(title_opts=opts.TitleOpts(title="生产总值",pos_top="15%"),
                     legend_opts=opts.LegendOpts(pos_top="10%"),
                     xaxis_opts=opts.AxisOpts(axislabel_opts={"rotate":20}))
line.render_notebook()
```

```python
#导入模块
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Pie
%matplotlib inline

#读取数据
data=pd.read_excel('本市去年部分统计数据.xls',sheet_name='县区指标2',skiprows=2)
#提取标签
items=list(data['固定资产投资增长速度（%）'][0:4])
#提取数据
y_data1=list(data['1-10月'][0:4])
#调用Pie()方法绘制饼图
c=Pie()
c.add("",[list(z) for z in zip(items,y_data1)])
c.set_global_opts(title_opts=opts.TitleOpts(title="各县(市、区、特区)主要经济指标"))
c.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c}"))
c.render_notebook()
```

```python
#导入模块
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Kline
%matplotlib inline

#读取数据
data=pd.read_excel('本市去年部分统计数据.xls',sheet_name='逐月情况',skiprows=2)
#提取y轴数据
y_data1=list(data['2月'][0:])
y_data2=list(data['3月'][0:])
y_data3=list(data['4月'][0:])
y_data4=list(data['5月'][0:])
y_data5=list(data['6月'][0:])
y_data6=list(data['7月'][0:])
y_data7=list(data['8月'][0:])
y_data8=list(data['9月'][0:])
y_data9=list(data['10月'][0:])
#进一步处理数据
datas=list([y_data1,y_data2,y_data3,y_data4,y_data5,y_data6,y_data7,y_data8,y_data9])
c=Kline()
c.add_xaxis(["2020/{}月".format(i+1) for i in range(12)])
c.add_yaxis("2020年逐月情况K线图",datas)
c.set_global_opts(yaxis_opts=opts.AxisOpts(is_scale=True),
                 xaxis_opts=opts.AxisOpts(is_scale=True),
                 title_opts=opts.TitleOpts(title="逐月情况"))
c.render_notebook()
```

```python
#导入模块
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Gauge,Page
%matplotlib inline

#读取数据
data=pd.read_excel('本市去年部分统计数据.xls',sheet_name='金融',skiprows=2)
#提取数据
y_data1=float(data['同比增长(%)'][0:1])
c=Gauge()
c.add("金融",[("同比增长(%)",y_data1)],
      axisline_opts=opts.AxisLineOpts(
          linestyle_opts=opts.LineStyleOpts(
              color=[(0.3,"#67e0e3"),
                     (0.7,"#37a2da"),
                     (1,"#fd666d")],width=30)))
c.set_global_opts(title_opts=opts.TitleOpts(title="金融"),
                  legend_opts=opts.LegendOpts(is_show=False))
c.render_notebook()
```

```python
#导入模块
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Page,WordCloud
from pyecharts.globals import SymbolType

#读取数据
data=pd.read_excel('本市去年部分统计数据.xls',sheet_name='县区指标',skiprows=2)
#提取标签
word=list(data['地区生产总值（亿元）'][0:4])
word1=word[0]
word2=word[1]
word3=word[2]
word4=word[3]
words=[(word1,5),(word2,3),(word3,8),(word4,2)]
c=WordCloud()
c.add("",words,word_size_range=[20,80])
c.set_global_opts(title_opts=opts.TitleOpts(title="县区指标"))
c.render_notebook()
```
