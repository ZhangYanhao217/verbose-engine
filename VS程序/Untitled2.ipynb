```python
import pymysql.cursors

# 连接数据库
connect = pymysql.Connect(
    host='10.67.15.106',
    port=3306,
    user='root',
    passwd='123456',
    db='titainic',
    charset='utf8'
)

# 获取游标
cursor = connect.cursor() 
cursor.execute("select sex from titanic")
result1=cursor.fetchall() #result1为一个元组
#将元组数据存进列表中
sex=[]

for row in result1:
    sex.append(row[0])

#统计词频
wordcount = {}
for word in sex:
    wordcount[word] = wordcount.get(word, 0)+1
list = sorted(wordcount.items(), key=lambda x: x[1], reverse=True)[0:]
x1 = [i[0] for i in list]
y1 = [i[1] for i in list]
print(x1)
print(y1)

#绘制玫瑰图
from pyecharts import options as opts
#绘制圆形饼图
from pyecharts.charts import Pie
pie = Pie() #init_opts = opts.InitOpts(theme = ThemeType.INFOGRAPHIC)
pie.add("",[(z) for z in zip(x1,y1)],radius=["40%","55%"],rosetype="radius")
pie.set_global_opts(title_opts=opts.TitleOpts(title="船上人数男女所占比例"))
pie.render_notebook()



cursor.execute("select Age from titanic")
result1=cursor.fetchall() #result1为一个元组
#将元组数据存进列表中
Age=[]

for row in result1:
    Age.append(row[0])

#统计词频
wordcount = {}
for word in Age:
    wordcount[word] = wordcount.get(word, 0)+1
list = sorted(wordcount.items(), key=lambda x: x[1], reverse=True)[1:20]
g = [i[0] for i in list]
h = [i[1] for i in list]
print(g)
print(h)

from pyecharts import options as opts
#绘制漏斗图
from pyecharts.charts import Funnel
ws =(Funnel()
      .add("该船上部分年龄段的人数分布",[(z) for z in zip(g,h)],is_selected = True))
ws.render_notebook()



cursor.execute("select Survived from titanic")
result1=cursor.fetchall() #result1为一个元组
#将元组数据存进列表中
Survived =[]

for row in result1:
    Survived.append(row[0])

#统计词频
wordcount = {}
for word in Survived:
    wordcount[word] = wordcount.get(word, 0)+1
list = sorted(wordcount.items(), key=lambda x: x[1], reverse=True)[0:]
a = [i[0] for i in list]
b = [i[1] for i in list]
print(a)
print(b)

#绘图开始
from pyecharts import options as opts
#绘制柱状图
from pyecharts.charts import Bar
bar = (Bar()
      .add_xaxis(a)
      .add_yaxis("0代表死亡，1代表生还",b)
      .set_global_opts(title_opts = opts.TitleOpts(title = "遇难人数")))
bar.render_notebook()
#bar.render()生成html

cursor.execute("select Pclass from titanic")
result1=cursor.fetchall() #result1为一个元组
#将元组数据存进列表中
Pclass =[]
for row in result1:
    Pclass.append(row[0])

#统计词频
wordcount = {}
for word in Pclass:
    wordcount[word] = wordcount.get(word, 0)+1
list = sorted(wordcount.items(), key=lambda x: x[1], reverse=True)[0:]
m = [i[0] for i in list]
n = [i[1] for i in list]
print(m)
print(n)

#绘图开始
from pyecharts import options as opts
#绘制圆形饼图
from pyecharts.charts import Pie
L1 = m
num = n
c = Pie()
c.add("",[(z) for z in zip(L1,num)])
c.set_global_opts(title_opts = opts.TitleOpts(title = "Pie - 船舱等级"))
c.set_series_opts(label_opts = opts.LabelOpts(formatter = "{b}:{c}"))
c.render_notebook()

cursor.execute("select Fare from titanic")
result1=cursor.fetchall() #result1为一个元组
#将元组数据存进列表中
Fare=[]
for row in result1:
    Fare.append(row[0])
    
#统计词频
wordcount = {}
for word in Fare:
    wordcount[word] = wordcount.get(word, 0)+1
list = sorted(wordcount.items(), key=lambda x: x[1], reverse=True)[0:15]
c = [i[0] for i in list]
d = [i[1] for i in list]
print(c)
print(d)    
#绘图开始
from pyecharts import options as opts
#水平条形图
from pyecharts.charts import Bar

bar = Bar()
bar.add_xaxis(c)
bar.add_yaxis("同等价位票数的显示",d)
bar.set_global_opts(title_opts = opts.TitleOpts(title = "每个票价对应的人数"),
                    toolbox_opts = opts.ToolboxOpts(is_show = True))
bar.set_series_opts(label_opts = opts. LabelOpts(position= "right"))
bar.reversal_axis()
bar.render_notebook()
```
